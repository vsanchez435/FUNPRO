```python
cero=0
n=int(input("Ingrese la cantidad de números a analizar: "))
for i in range(1,n+1):
  num=int(input(f"Ingrese el numero {i}: "))
  if num==0: cero=cero+1
print(f"La cantidad de ceros en {n} números es {cero} ")
```

```python
#promedio de pares y suma de impares
#50 números cualesquiera
impar=0
pares=0
n=0
for i in range(10):#Quiero que se repita una cantidad de veces, en este caso, 10
  num=int(input("Ingrese un número: "))
  if num%2==0:
    pares=pares+num
    n=n+1
  else: impar=impar+num

print(f"La suma de los números impares es {impar}")
print(f"El promedio de los números pares es {pares/n}")
```
